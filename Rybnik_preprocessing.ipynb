```python
#### Ładowanie Danych Z rozpakowanego Folderu.
#Na początku Procesowane są dane dotyczące korzystania z Kart Miejskich


import pandas as pd
import os


path = 'C:\\Users\\ldoma\\Downloads\\Rybnik\\'


def skasowan_karty(path):
    
    column_names = ['Liczba_porządkowa'
                    ,'Data_operacji'
                    ,'Numer_karty'
                    ,'Numer_boczny_pojazdu'
                    ,'Numer_linii'
                    ,'Wariant_trasy'
                    ,'Oznaczenie_kursu'
                    ,'Pierwszy_przystanek'
                    ,'Drugi_przystanek'
                    ,'Numer_ewidencyjny'
                    ,'Typ_operacji'
                    ]
    
    path_2 = path+'analiza-skasowane-kart_2021-11-19_0624\\analiza skasowańe-kart\\'
    file_list = [f for f in os.listdir(path_2) if os.path.isfile(os.path.join(path_2, f))]
    df_temp_1 =  pd.read_csv(path_2+str(file_list[0]),delimiter =';', header=None)
    for i in file_list[1:]:
        df_temp = pd.read_csv(path_2+str(i),delimiter =';', header=None)
        df_temp_1 = pd.concat([df_temp_1,df_temp])
    df_temp_1.reset_index(inplace=True)
    df_temp_1.drop(['index',11],axis=1,inplace=True)
    for i in df_temp_1.columns:
        df_temp_1.rename(columns={i:column_names[i]},inplace=True)
    df_temp_1.drop(['Liczba_porządkowa'],axis=1,inplace=True)
    
    return df_temp_1

df_karty = skasowan_karty(path)
df_karty
```

```python
#Usuwanie Nullów, po ustaleniu że stanowią one niecałe 0.5 promila danych

df_karty = df_karty.dropna()
```

```python
## Wyciągnięcie z Oznaczenia kursu Numeru Kursu oraz godzinę przyjazdu, po czym zamiana jej na odpowiedni format.

import datetime as dt

df_karty['Godzina_przyjazdu'] = df_karty['Oznaczenie_kursu'].apply(lambda str_test : dt.datetime.strptime(str(str_test)[-6:-1], '%H:%M').time())
df_karty['numer_kursu'] = df_karty['Oznaczenie_kursu'].apply(lambda str_test : str(str_test)[:-7])
df_karty.drop('Oznaczenie_kursu',axis=1,inplace=True)
df_karty
```

```python
#Dodanie Daty i Godziny razem w typie Datetime

df_karty_przepracowane_bez_okresowych_ZWRTiNIE['data'] = df_karty_przepracowane_bez_okresowych_ZWRTiNIE['Data_operacji']+ pd.to_timedelta(df_karty_przepracowane_bez_okresowych_ZWRTiNIE['Godzina_przyjazdu'].astype(str))
```

```python
# Pracowanie po kolei Recordów Wedle Pobrania Według Daty --> grupując Cel i Początek razem
# Tworzymy do tego coelu nową DF za każdum razem dodając do listy tablic, po czym konkatenujemy je 
# Zgodnie z Zasadą że Pierwszy dla Pobrania oznacza początek trasy, a Drugi dla zwrotu oznacza Koniec trasy


list_of_cards = df_karty_przepracowane_bez_okresowych_ZWRTiNIE['Numer_karty'].unique()
temp_ost_df = []
for i in list_of_cards:
    temp_df = df_karty_przepracowane_bez_okresowych_ZWRTiNIE[df_karty_przepracowane_bez_okresowych_ZWRTiNIE['Numer_karty'] == i]
    list_of_hours = temp_df['data'].unique()
    
    temp_dfs = []
    #print(i)
    count = 1
    for j in list_of_hours:
        #print(j)
        
        
        try:
            temp_first_stop = list(temp_df[(temp_df['data'] == j) & (temp_df['pobranie '] == 'Pobranie')]['Pierwszy_przystanek'])[0]
        except:
            temp_first_stop =np.nan
        try:
            temp_last_stop = list(temp_df[(temp_df['data'] == j) & (temp_df['pobranie '] == 'Zwrot')]['Drugi_przystanek'])[0]
        except:
            temp_last_stop = np.nan
        new_temp_df = temp_df[(temp_df['data'] == j) & (temp_df['pobranie '] == 'Pobranie')]
        new_temp_df['first_stop'] = temp_first_stop
        new_temp_df['last_stop'] = temp_last_stop

        new_temp_df.drop(['Data_operacji','Godzina_przyjazdu','Pierwszy_przystanek','Drugi_przystanek','pobranie '],axis=1,inplace=True)
        temp_dfs.append(new_temp_df)
        count +=1
        
        if count > len(list_of_hours):
            break
    temp_ost_df.append(pd.concat(temp_dfs))
pd.concat(temp_ost_df)
```

```python
DRUGI_DF = pd.concat(temp_ost_df)
DRUGI_DF_2 = DRUGI_DF.dropna()
DRUGI_DF_2.to_csv(path + 'Preprocessing_1.csv', index = False)
DRUGI_DF_2
```
