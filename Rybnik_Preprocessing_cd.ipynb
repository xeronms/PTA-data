```python
##Ciąg Dalszy Preoricessingu: 
#Tutaj mamy prostą funckję którea pozwala zeksplorować liczbę ruchu na linii 

import pandas as pd
import os
import datetime as dt

path = 'C:\\Users\\ldoma\\Downloads\\Rybnik\\'

DRUGI_DF_2 = pd.read_csv(path+'Preprocessing_1.csv')


dane = DRUGI_DF_2
def liczba_ruchu_na_linii(first,last):
    
    return len(dane[(dane['first_stop'] == str(first)) & (dane['last_stop'] == str(last))])

liczba_ruchu_na_linii('Now Maroko' ,'� Dworzec Kolej')
```

```python
#Kod, któr przez input zmienia nazyw przystanków na Ładniejsze

dane2 = dane
przystanki = dane2['last_stop'].unique()
przystanki.sort()
for i in przystanki:
    new_string = input(i) 
    dane2['last_stop'] = dane2['last_stop'].apply(lambda j: new_string if  j==i else j)
    dane2['first_stop'] = dane2['first_stop'].apply(lambda j: new_string if  j==i else j)
```

```python
dane2.to_csv(path + 'Preprocessing_2.csv', index = False)
```

```python
#Rozdzielenie Daty i Godziny oraz Wydzielenie  Dnia Tygodnia

dane2['data'] = pd.to_datetime(dane2['data'])
dane2['Dzień_tyg'] = dane2['data'].apply(lambda i : i.weekday())
dane2['Godzina'] = dane2['data'].apply(lambda tdate : dt.datetime.strptime("{:d}:{:02d}".format(tdate.hour, tdate.minute), '%H:%M').time())
dane2
```

```python
Przystanki = Przystanki[Przystanki[' GPS-D�ugo�� '] != 0]
Przystanki
```

```python
Lista_przystanków = Przystanki[' Nazwa przystanku '].unique()
```

```python
#Obliczenie i pRzupisanie  po czym usunięcie Dublujących sie przystanków. Wzięcie Koordynatów przystanków powielonych jako średnią ich wspórzędnych
#

list_of_tempss = []
for i in Lista_przystanków:
    temp_df = Przystanki[Przystanki[' Nazwa przystanku '] == i]
    temp_df[' GPS-D�ugo�� '] = temp_df[' GPS-D�ugo�� '].mean()
    list_of_tempss.append(temp_df)
df_after_prerp = pd.concat(list_of_tempss)
```

```python
df_after_prerp
result_df = df_after_prerp.drop_duplicates(subset=[' Nazwa przystanku '], keep='first')
result_df.reset_index(inplace=True)
result_df.drop('index',axis=1,inplace=True)
result_df
```

```python
#Wyczyszczenie i Zapisanie pliku Geograficznego do dalszego uzytkowania

result_df['name'] = result_df[' Nazwa przystanku ']
result_df['id'] = result_df[' Nrprzystanku ']
result_df['longitutde'] = result_df[' GPS-D�ugo�� ']
result_df['latitude'] = result_df[' GPS-Szeroko�� ']
result_df.drop([' Nazwa przystanku ',' Nrprzystanku ', ' GPS-D�ugo�� ',' GPS-Szeroko�� '],axis=1,inplace=True)
```
